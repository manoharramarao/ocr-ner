```python
import matplotlib.pyplot as plt
import keras_ocr
import os
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
custom_images = []
custom_images.append(os.path.join(
    '/home/manohar/projects/personal/invoice_scanner/ocr_1', 'test4.jpg'))
custom_images.append(os.path.join(
    '/home/manohar/projects/personal/invoice_scanner/ocr_1', 'test.jpg'))
```

```python
images = [ keras_ocr.tools.read(path) for path in custom_images]
```

```python
%time predictions = pipeline.recognize(images)
# print(predictions)
```

```python
# fig, axs = plt.subplots(nrows=len(images), figsize=(10, 10))
# if(len(images) == 1):
#   for image, prediction in zip(images, predictions):
#     keras_ocr.tools.drawAnnotations(image=image, predictions=prediction, ax=axs)
# else:
#   for ax, image, prediction in zip(axs, images, predictions):
#     keras_ocr.tools.drawAnnotations(image=image, predictions=prediction, ax=ax)
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, predictions):
    keras_ocr.tools.drawAnnotations(
        image=image, predictions=predictions, ax=ax)
```

```python
with open('results.txt', 'a+') as f:
  for idx, prediction in enumerate(predictions):
    if(idx != 0):
      print("\n")
      f.write("\n\n")
    print("Results for the file: " + os.path.basename(custom_images[idx]))
    f.write("Results for the file: " +
            os.path.basename(custom_images[idx]) + ":\n\n")
    for word, array in prediction:
      if word == "\n":
        print("\n")
        f.write("\n")
      else:
        print(word,  end=' ')
        f.write(word + " ")
```
